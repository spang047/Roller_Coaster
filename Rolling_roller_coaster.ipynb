```python
import pandas as pd
import matplotlib.pyplot as plt

# load rankings data here:
ranking = pd.read_csv('roller_coasters.csv')
print(ranking.head())
#print(ranking.status.unique())
#wood_coaster = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
print(wood_coaster.head())
steel_coaster = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')
#print(steel_coaster.head())

               
# write function to plot rankings over time for 1 roller coaster here:
    #take in array of values and the name of a roller coaster to plot values
def ranking_one_coaster(df_name, coaster_name):
    x_values = df_name['Year of Rank'][(df_name.Name == coaster_name)]
    y_values = df_name.Rank[(df_name.Name == coaster_name)]
    ax = plt.subplot()
    plt.plot(x_values, y_values)
    plt.title('Rank over Time')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.legend([coaster_name])
    plt.show()
    plt.clf()
    
ranking_one_coaster(steel_coaster, 'Nitro')
```

```python
def ranking_two_coaster(df_name, coaster_name1, coaster_name2):
    x_values = df_name['Year of Rank'][(df_name.Name == coaster_name1)]
    y_values1 = df_name.Rank[(df_name.Name == coaster_name1)]
    y_values2 = df_name.Rank[(df_name.Name == coaster_name2)]
    ax = plt.subplot()
    plt.plot(x_values, y_values1)
    plt.plot(x_values, y_values2)
    plt.title('Rank over Time')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.legend([coaster_name1, coaster_name2])
    plt.show()
    plt.clf()

ranking_two_coaster(steel_coaster,'Apolloâs Chariot', 'Millennium Force')
```

```python
# write function to plot top n rankings over time here:
def ranking_n_coaster(rankings_df, n):
    top_n_rankings = rankings_df[rankings_df['Rank'] <= n]
    ax = plt.subplot()
    for coaster in set(top_n_rankings['Name']):
      coaster_rankings = top_n_rankings[top_n_rankings['Name'] == coaster]
      ax.plot(coaster_rankings['Year of Rank'],coaster_rankings['Rank'],label=coaster)
    plt.title('Rank over Time')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    #plt.legend(array_names)
    plt.show()
    plt.clf()

ranking_n_coaster(steel_coaster, 5)
```

```python
# write function to plot histogram of column values here:
def column_hist(df_name, column_name):
    plt.hist(df_name[column_name], bins = 40)
    plt.xlabel(column_name)
    plt.ylabel('Values')
    plt.title('Histogram of ' + column_name)
    plt.show()
    plt.clf()

column_hist(ranking, 'length')
```

```python
# write function to plot inversions by coaster at a park here:

def inversion_plot(df_name,park_name):
    inversions = df_name.num_inversions[df_name.park == park_name]
    coaster_name = df_name.name[df_name.park == park_name]
    ax = plt.subplot()
    plt.plot(coaster_name, inversions)
    plt.xlabel('Coaster Names')
    plt.ylabel('Inversions')
    plt.title('Number of inversions by coaster at ' + park_name)
    ax.set_xticks(range(len(coaster_name)))
    ax.set_xticklabels(coaster_name, rotation = 45)
    plt.show()
    plt.clf()


inversion_plot(ranking, 'Le Pal')


plt.clf()
```

```python
# write function to plot pie chart of operating status here:
def operating_pie(dataframe):
    statuses = ['status.operating', 
                'status.relocated', 
                'status.closed.definitely', 
                'status.retracked', 
                'status.closed.temporarily',
                'status.unknown',
                'status.construction',
                'status.announced', 
                'status.rumored']
    simple_statuses =  ['operating', 
                'relocated', 
                'closed.definitely', 
                'retracked', 
                'closed.temporarily',
                'unknown',
                'construction',
                'announced', 
                'rumored']
    counts = []
    for i in statuses:
        num = dataframe.status[dataframe.status == i].count()
        counts.append(num)
   
    plt.figure(figsize=(7,7))
    plt.pie(counts, autopct='%1.0f%%', shadow=True, pctdistance=.85)
    plt.axis('equal')
    plt.title('% of Operating and Non-Operating')
    plt.legend(simple_statuses)
    plt.show()
    plt.clf()





operating_pie(ranking)
plt.clf()
```

```python
# write function to create scatter plot of any two numeric columns here:
def two_scatter(df_name, column1, column2):
    plt.scatter(x = df_name[column1], y = df_name[column2])
    plt.xlabel(str(column1))
    plt.ylabel(str(column2))
    plt.title(str(column1) + ' vs. ' + str(column2))
    plt.show()
    plt.clf()



two_scatter(steel_coaster, 'Year Built', 'Points')




plt.clf()
```
